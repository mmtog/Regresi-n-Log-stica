```python
#Cargamos librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
grecia =pd.read_csv("Grecia_limpios_final.csv")
grecia.head(5)
```

```python
# Reemplazar 'unknown' por NaN
grecia['bathrooms'] = grecia['bathrooms'].replace('Unknown', np.nan)
grecia['bedrooms'] = grecia['bedrooms'].replace('Unknown', np.nan)
# convertir a float
grecia['bathrooms'] = grecia['bathrooms'].astype(float)
grecia['bedrooms'] = grecia['bedrooms'].astype(float)
```

```python
#Rellenamos valores nulos
grecia= grecia.fillna(method='bfill')
greciaf= grecia.fillna(method='ffill')
```

```python
grecia.info()
```

```python
grecia['minimum_nights'].info()
```

#** Regre Log 1 **



```python
#Obtenemos el limite superior y el límite inferior de la columna objetivo
Max=grecia['minimum_nights'].max()
Min=grecia['minimum_nights'].min()
Limites= [Min, Max]
Limites
```

```python
#Categorización de variables
#Declaramos 2 intervalos
intervalos=np.linspace(.99, 3.1, 3)
intervalos
```

```python
#Creamos las categorías
categorias= ["visit", "vacation"]
```

```python
#Finalmente creamos las categorías en la columna numérica
grecia['minimum_nights']=pd.cut(x= grecia['minimum_nights'], bins=intervalos, labels= categorias )
```

```python
#Rellenamos valores nulos
grecia= grecia.fillna(method='bfill')
greciaf= grecia.fillna(method='ffill')
```

```python
grecia['minimum_nights'].isnull().sum()
```

```python
nulos= grecia.isnull().sum().sum()
nulos
```

```python
vars_Indep = grecia[['bathrooms']]
vars_Dep = grecia ['minimum_nights']
```

```python
#Redefinimos las variables
x= vars_Indep
y= vars_Dep
```

```python
#dividimos el conjunto de datos en la parte de entrenamiento y pruba
X_test, X_train, Y_test, Y_train= train_test_split(x,y, test_size=0.3, random_state=None)
```

```python
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba, Utilizaremos
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, Y_train)
```

```python
#realizamos una prediccion
y_pred = algoritmo.predict(X_test)
y_pred
```

```python
#Verifico la matriz de confusion
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(Y_test, y_pred)
print('Matriz de Confusion')
print(matriz)
```

```python
#calculamos la prediccion del modelo
from sklearn.metrics import precision_score

precision = precision_score(Y_test, y_pred, average="binary", pos_label="visit")
print('Precision del modelo:')
print(precision)
```

```python
#Calculamos la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(Y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```

```python
#calculamos la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad = recall_score(Y_test, y_pred, average="binary", pos_label="visit")
print('Sensibilidad del modelo:')
print(sensibilidad)
```
